## Import the relevant libraries

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from sklearn.cluster import KMeans
from sklearn import preprocessing
```

## Load the data

Load data from the csv file: <i> 'iris_dataset.csv'</i>.

```python
data = pd.read_csv('iris-dataset.csv')

data
```

## Plot the data

For this exercise, try to cluster the iris flowers by the shape of their sepal. 

<i> Use the 'sepal_length' and 'sepal_width' variables.</i> 

```python
#plt scatter
plt.scatter(data['sepal_length'],data['sepal_width'])
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

# Clustering (unscaled data)

Separate the original data into 2 clusters.

```python
#select your desired features (widht,length) from the dataset
#x = data.iloc[:,0:2]
#x
```

```python
# create a variable which will contain the data for the clustering
x = data.copy()
```

## Now lets do the actual clustering

```python
#create a cluster object. We want 2 in this case
kmeans = KMeans(2)

#now fit the cluster into our selected data
kmeans.fit(x)
```

Now our selected features are fitted into 2 clusters

## Clustering results

now we got to predit the clusters

```python
identified_clusters = kmeans.fit_predict(x)
identified_clusters
```

**Now we want to copy our original dataset into a dataset with the identified clusters**

```python
data_w_clusters = data.copy()

#add cluster column to data_w_clusters
data_w_clusters['Identified Clusters'] = identified_clusters

#now check dataset
data_w_clusters
```

## Plot the data

```python
#plt scatter

plt.scatter(data['sepal_length'],data['sepal_width'], c=data_w_clusters['Identified Clusters'], cmap='rainbow')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

# Standardize the variables

Import and use the <i> method </i> function from sklearn to standardize the data. 

```python
# Scale the inputs
# preprocessing.scale scales each variable (column in x) with respect to itself
# The new result is an array. Lets create a new variable that has a preprocessed x
x_scaled = preprocessing.scale(data)
x_scaled
```

# Clustering (scaled data)

```python
#so now lets repeat the clustering steps, BUT THIS TIME with x_scaled NOT x

#create clustering object (2)
kmeans_new = KMeans(2)

#fit the data
kmeans_new.fit(x_scaled)
```

```python
# Create a new data frame with the predicted clusters

#lets copy the original dataset
clusters_scaled = data.copy()

#lets add the predicted values to clusters_new
clusters_scaled['clusters_pred'] = kmeans_new.fit_predict(x_scaled)

#check 
clusters_scaled
```

```python
#Now lets plot

plt.scatter(clusters_scaled['sepal_length'],clusters_scaled['sepal_width'],c=clusters_scaled['clusters_pred'], cmap='rainbow')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

**RESULTS**

Looks like the two solutions are identical. That is because the original features have very similar scales to start with!

## Take Advantage of the Elbow Method

### WCSS

```python
# Createa an empty list
wcss= []

# Create all possible cluster solutions with a loop
# We have chosen to get solutions from 1 to 9 clusters; you can ammend that if you wish
for i in range(1,10):
    # Clsuter solution with i clusters
    kmeans = KMeans(i)
    # Fit the STANDARDIZED data
    kmeans.fit(x_scaled)
    # Append the WCSS for the iteration
    wcss.append(kmeans.inertia_)
    
# Check the result
wcss
```

### The Elbow Method

```python
#lets plot this bad boy. Plot the number of clusters vs WCSS
plt.plot(range(1,10),wcss)
# Name your axes
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

How many clusters are there?

2 to 3
